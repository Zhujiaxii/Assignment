```python
# Actually i tried to write another programme to click the bottom botton automatically. But I failed. 
# The mistaken code is in the file [Wrong Code]. It will be nice if you could tell me how to correct it. Thank you!


import urllib
import requests
import numpy as np
from bs4 import BeautifulSoup
import csv
import pandas as pd
import os
import wget
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time



header = ['page #','title','introduction','url']
data = []


for i in range(1,2):
    session = requests.Session()
    headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"}
    req = session.get("https://www.douban.com/search?cat=1015&q=香港电影取景地",headers = headers)
    soup = BeautifulSoup (req.content)
    
    #print(soup)
    print("\n***** Page " + str(i) +" in action *****")
    
    for article in soup.find_all("div", attrs={"class":"result"}):
        page_no = str(i) 
        title = article.find("div",attrs={"class":"title"}).get_text()
        introduction = article.p.get_text()
        url = article.a.get('href')
        print(title)
        print(introduction)
        print(url)
        data.append((page_no, title, introduction, url))
        
df = pd.DataFrame(data,
  columns = header
)
df.to_csv('douban_data_2.csv', sep='\t', encoding='utf-8')
df
```
